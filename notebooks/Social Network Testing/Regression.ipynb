```python
# Import necessary libraries
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
import pandas as pd
```

```python
# DIRECTORIES
DATA = '/ais/hal9000/datasets/reddit/stance_pipeline/luo_data/network_analysis/statistics/'
```

```python
# Load data
TOLERANCE = 100
data = pd.read_csv(DATA+f"2019_NX_T{TOLERANCE}.csv")
```

```python
# Regression
X = np.array(data['Size']).reshape(-1,1)
Y = np.array(data["Loyalty"]).reshape(-1,1)

model = LinearRegression()
model.fit(X, Y)

coefficients = model.coef_
intercept = model.intercept_

print("Coefficients:", coefficients)
print("Intercept:", intercept)
```

```python
X = np.log(np.array(data['Size'])).reshape(-1,1)
Y = np.array(data["Connectivity"]).reshape(-1,1)

plt.scatter(X, Y, alpha=0.5, label='Data')
plt.xlabel('X')
plt.ylabel('y')
plt.show()
```
